```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_ = pd.read_csv('data_without_outliers.csv')
data5_ = pd.read_csv('cleaned_data.csv')
```

```python
categorical_columns = data_.select_dtypes(include=['object']).columns
numeric_columns = data_.select_dtypes(include=['float64',]).columns
```

```python
# Histograms for numerical features
df_numerical = data5_[numeric_columns]
df_numerical.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Bar charts for categorical features
plt.figure(figsize=(10, 6))
sns.countplot(x=data5_['District'])
plt.title(f'Distribution of District')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar charts for categorical features
plt.figure(figsize=(10, 6))
sns.countplot(x=data5_['Cluster'])
plt.title(f'Distribution of Cluster')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of target variable
plt.figure(figsize=(10, 6))
sns.countplot(x=data5_['ProgressStatus'])
plt.title('Distribution of ProgressStatus')
plt.show()
```
